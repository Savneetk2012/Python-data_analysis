```python
import pandas as pd

ht=pd.read_csv("C:/Users/savneet kuar/Documents/datasets/heart.csv")
ht
```

```python
ht.head()
```

```python
df=ht.drop(columns=['thal','target','ca','slope'])
df
```

```python
#age in years
#trestbps - resting blood pressure (in mmHg on admission to the hospital)
#sex = 1-M 0-F
#cp- chest pain type
#fbs- fasting blood sugar
#restecg- resting electrocardiographic results
#oldpeak- ST depression induced by exercise relative to rest
#thalach- maximum heart rate achieved
#exang- exercise induced angina
```

```python
mht=df[df["sex"]==1]
mht
```

```python
fht=df[df["sex"]==0]
fht
```

```python

import matplotlib.pyplot as plt

#Heart attacks in males vs females as per age groups

plt.figure(figsize=(30,10))
plt.subplot(1,2,1)
plt.title('Heart attacks in females')
plt.bar(fht.age, fht.chol,color='r')


plt.subplot(1,2,2)
plt.title('Heart attacks in males')
plt.bar(mht.age,mht.chol)


plt.show()
```

```python
plt.figure(figsize=(30,10))
plt.subplot(1,2,1)
plt.title('Heart attacks in females')
plt.xlabel('Age of females')
plt.ylabel('Chest pain')
plt.bar( fht.age,fht.cp,color='r')


plt.subplot(1,2,2)
plt.title('Heart attacks in males')
plt.xlabel('Age of males')
plt.ylabel('chest pain')
plt.bar(mht.age,mht.cp,)


plt.show()
```

```python
plt.figure(figsize=(30,10))
plt.subplot(1,2,1)
plt.title('Heart attacks in females')
plt.xlabel('Age of females')
plt.ylabel('Resting blood pressure (in mmHg on admission to the hospital)')
plt.bar(fht.age, fht.trestbps,color='r')


plt.subplot(1,2,2)
plt.title('Heart attacks in males')
plt.xlabel('Age of males')
plt.ylabel('Resting blood pressure (in mmHg on admission to the hospital)')
plt.bar(mht.age,mht.trestbps)
```

```python
plt.figure(figsize=(30,10))
plt.subplot(1,2,1)
plt.title('Heart attacks in females')
plt.xlabel('Age of females')
plt.ylabel('Maximum heart rate achieved')
plt.bar(fht.age, fht.thalach ,color='r')


plt.subplot(1,2,2)
plt.title('Heart attacks in males')
plt.xlabel('Age of males')
plt.ylabel('Maximum heart rate achieved')
plt.bar(mht.age,mht.thalach)
```

```python


plt.figure(figsize=(30,10))
plt.subplot(1,2,1)
plt.title('Heart attacks in females')
plt.xlabel('Age of females')
plt.ylabel('ST depression induced by exercise relative to rest')
plt.bar(fht.age, fht.oldpeak,color='r')


plt.subplot(1,2,2)
plt.title('Heart attacks in males')
plt.xlabel('Age of males')
plt.ylabel('ST depression induced by exercise relative to rest')
plt.bar(mht.age,mht.oldpeak)

plt.show()
```

CONCLUSION
Males falling under the age groups of 60-70 and females falling under the age groups 50-60 are more likely to suffer  a heart attack as compared to other age groups of their respective sex owing to the risk factors as shown in the graphs.
